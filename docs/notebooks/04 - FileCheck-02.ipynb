# File Check

To make sure data fetched from external databases is fully available and in the right directories, a full file check ensures later processes do not halt.
The file check works as follows
- The user adjusts parameter values in the user variables section
- Two nested loops count generate filenames with paths for each month of data
- In the nested loops, filePathList and fileExists are two lists that are appended accordingly to if a file is found or not
- Similar process is duplicated for each dataset
- At the end, both lists (filePathList, fileExists) are added to the fileCheckDF pandas dataframe
- The created pandas dataframe is exported to a file
- The dataframe is also used to visualized to see how much and which month's data is missing

## 1. Setup

```python
# Imported Libraries
import os
import pandas as pd
```

```python
# User Variables & Parameters
# Control Parameters
prec_FileName           = 'prec_'
tmin_FileName           = 'tmin_'
tmax_FileName           = 'tmax_'
prec_DataPath           = '../../WaterBucket/data/isb/prec/'
tmin_DataPath           = '../../WaterBucket/data/isb/tmin/'
tmax_DataPath           = '../../WaterBucket/data/isb/tmax/'
startYear               = 1961
endYear                 = 2018
numYears                = (endYear - startYear) + 1
fromMonth               = 1
toMonth                 = 12

filePathList           = []
fileExists             = []
fileCheckDF             = pd.DataFrame()
```

## 2. File Check

```python
# WorldClim Precipitation Data
for y in range(startYear, endYear + 1):
    for m in range(fromMonth, toMonth + 1):

        # Handling case where month name is 01, 02, to 09
        monthName = ''
        if m < 10:
            monthName = '0' + str(m)
        else:
            monthName = str(m)
        
        # Calculaing imagepath and searching if it exists on the system
        fileName = prec_FileName + str(y) + '-' + str(monthName) + '.tif'
        imagePath = prec_DataPath + fileName
        filePathList.append(imagePath)
        
        if (os.path.exists(imagePath)):
            print('Found File:\t', fileName)
            fileExists.append(1)
        else:
            print('Missing File:\t', fileName)
            fileExists.append(0)
```

```python
# WorldClim Minimum Temperature Data
for y in range(startYear, endYear + 1):
    for m in range(fromMonth, toMonth + 1):

        # Handling case where month name is 01, 02, to 09
        monthName = ''
        if m < 10:
            monthName = '0' + str(m)
        else:
            monthName = str(m)
        
        # Calculaing imagepath and searching if it exists on the system
        fileName = tmin_FileName + str(y) + '-' + str(monthName) + '.tif'
        imagePath = tmin_DataPath + fileName
        filePathList.append(imagePath)
        
        if (os.path.exists(imagePath)):
            print('Found File:\t', fileName)
            fileExists.append(1)
        else:
            print('Missing File:\t', fileName)
            fileExists.append(0)
```

```python
# WorldClim Maximum Temperature Data
for y in range(startYear, endYear + 1):
    for m in range(fromMonth, toMonth + 1):

        # Handling case where month name is 01, 02, to 09
        monthName = ''
        if m < 10:
            monthName = '0' + str(m)
        else:
            monthName = str(m)
        
        # Calculaing imagepath and searching if it exists on the system
        fileName = tmax_FileName + str(y) + '-' + str(monthName) + '.tif'
        imagePath = tmax_DataPath + fileName
        filePathList.append(imagePath)
        
        if (os.path.exists(imagePath)):
            print('Found File:\t', imagePath)
            fileExists.append(1)
        else:
            print('Missing File:\t', imagePath)
            fileExists.append(0)
```

```python
# Making a file report
fileCheckDF['filePath']     = filePathList
fileCheckDF['fileExists']   = fileExists
```
